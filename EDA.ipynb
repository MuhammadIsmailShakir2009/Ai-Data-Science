# I obtained the dataset from Kaggle. Since the data was already clean, I did not perform additional cleaning or feature engineering. My primary focus in this project was on data visualization, where I explored patterns and insights using various charts and graphs

```python
import seaborn as sns        # seaborn
import pandas as pd          # pandas
import numpy as np           # numpy
import math                  # math
```

```python
data_1= pd.read_csv('taxis_cleaned.csv')
```

```python
data_1
```

```python
from matplotlib import pyplot as plt
data_1['tip'].plot(kind='hist', bins=20, title='tip')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
data_1.plot(kind='scatter', x='fare', y='tip', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data_1.plot(kind='scatter', x='tip', y='tolls', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
data_1['tip'].plot(kind='line', figsize=(8, 4), title='tip')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
data_1.plot(kind='scatter', x='tip', y='tolls', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['dropoff_borough'].value_counts()
    for x_label, grp in data_1.groupby('pickup_borough')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('pickup_borough')
_ = plt.ylabel('dropoff_borough')
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['pickup_borough'].value_counts()
    for x_label, grp in data_1.groupby('payment')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('payment')
_ = plt.ylabel('pickup_borough')
```

```python
figsize = (12, 1.2 * len(data_1['payment'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data_1, x='passengers', y='payment', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
figsize = (12, 1.2 * len(data_1['color'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data_1, x='passengers', y='color', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['payment'].value_counts()
    for x_label, grp in data_1.groupby('color')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('color')
_ = plt.ylabel('payment')
```
